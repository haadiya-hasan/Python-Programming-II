An **Iterator** in Python is an object that allows sequential access to elements of a collection (for eg: lists, tuples etc.) without exposing the underlying structure. Iterators provide a way to access elements one at a time and remember their state between iterations.

```python
# Creating a simple iterator using iter() and next()
num = [1, 2, 3, 4, 5]  # Iterable (list)
num_iter = iter(num)  # Creating an iterator
print(next(num_iter))
```

```python
# Creating a string and iterating it.

string = "Hello"
string_iter = iter(string)
print(next(string_iter))
print(next(string_iter))
```

```python
#Using a loop to execute the iteration
l = [1,2,3,4,5,6,7,8]
iterator = iter(l)
for item in iterator:
  print(item)
```

```python
#Using iteration for a range

range_iter = iter(range(5))
print(next(range_iter))
print(next(range_iter))
```

```python
#While loop
fruits = ("apple", "banana", "cherry")
fruit_iter = iter(fruits)

while True:
    try:
        print(next(fruit_iter))
    except StopIteration:
        break
```

```python
#The same program, but its for "for loop"

fruits = ("apple", "banana", "cherry")

for fruit in fruits:
    print(fruit)
```

```python
numbers = [10, 20, 30, 40, 50]
num_iter = iter(numbers)   # create iterator
while True:
    try:
        print(next(num_iter))
    except StopIteration:
        break
```

A **Decorator** in Python is a function which takes other functions as input, add additional functionalities and returns it.

PS : Think of a decorator as a wrapper: it takes a function, adds some kind of functionality, and then returns a new function with the added behavior.

```python
#With Decorator
def decor(func):
    def wrapper():

      print ("not start")
      func()
    return wrapper

@decor
def printer():
    print("Hello!")
    print("2Hello!")

# Directly calling the decorated function
printer()
```

```python
def decor(func):
  def wrapper():
    print('------')
    func()
    print('------')
  return wrapper

@decor
def printer():
  print('Hello')

printer()
```

```python
def my_decorator(func):
    def wrapper():
        print("Before the function runs")
        func()
        print("After the function runs")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()

# @ my_decorator is a shortcut for say_hello = my_decorator(say_hello)
```

```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

```python
#Timer Decorator
#2. Timer Decorator (No Arguments)
#Question: Create a decorator that measures execution time of a function with no parameters.

import time

def timer_decorator(func):
    def wrapper():
        start = time.time()
        result = func()
        end = time.time()
        print(f"Time taken: {end - start:.4f} seconds")
        return result
    return wrapper

@timer_decorator
def do_work():
    time.sleep(1)
    print("Work done.")

do_work()
```

```python
#3. Authorization Decorator
def requires_admin(func):
    def wrapper(user):
        if user != "admin":
            print("Access denied")
            return
        return func(user)
    return wrapper

@requires_admin
def access_dashboard(user):
    print(f"{user} accessed the dashboard.")

access_dashboard("guest")
access_dashboard("admin")
```

```python
#Factorial using Decorators
def decor(func):
    cache = {}
    def wrapper(n):
        if n in cache:
            return cache[n]
        result = func(n)
        cache[n] = result
        return result
    return wrapper

@decor
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

**Generators** in Python are a special type of iterable that allow you to iterate over data lazily, meaning they generate values on the fly instead of storing them in memory. This makes generators very memory-efficient for handling large data sets.

```python
#A generator function looks like a normal function but uses yield instead of return.

def my_gen():
    yield "sumit"  #keyword
    yield 2
    yield 3

gen = my_gen()  # Creating the generator
print(gen)
```

```python
def my_generator():
    yield 1  #keyword
    yield 2
    yield 3

gen = my_generator()  # Creating the generator

for num in gen:
    print(num)  # Output: 1, 2, 3
```

```python
#Generator for Fibonacci Series

def fibonacci(n):
    a, b = 0, 1
    for n in range(n):
        yield a
        a, b = b, a + b

fib = fibonacci(100)
print(list(fib))
```

```python
#Using send() to Interact with a Generator

def countdown(n):
    while n > 0:
        val = (yield n)
        if val is not None:
            n = val
        else:
            n -= 1

cd = countdown(5)
print(next(cd))
print(cd.send(3))
print(next(cd))
```
