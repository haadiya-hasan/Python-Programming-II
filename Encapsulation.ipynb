# 🔹 Encapsulation in Object-Oriented Programming (OOP)

Encapsulation is one of the fundamental concepts of OOP that refers to **wrapping data (variables) and methods (functions) together into a single unit (class)**.

It is mainly used to **restrict direct access to data** and protect it from unwanted modifications. This is achieved using **access modifiers** like private variables.

## 🔹 Key Points
- Combines data and methods into a single class
- Protects data using **private variables**
- Access to data is controlled using **public methods**
- Improves security and data integrity

## 🔹 In Python
- Private variables are defined using `__` (double underscore)
- Data is accessed using methods (getter/setter or display functions)

## 🔹 Real-Life Example
Think of a **bank account 💳**:
- Your balance is hidden (private)
- You can access it only through functions like check_balance()

👉 This is encapsulation — **data hiding with controlled access**

```python
class Person:

    # Constructor method → initializes object attributes
    def __init__(self, name, age):
        self.__name = name  # Private attribute (cannot be accessed directly)
        self.__age = age    # Private attribute

    # Public method (Accessor) to display private data
    def display_info(self):
        # Accessing private variables inside the class
        print(f"Name: {self.__name}, Age: {self.__age}")


# Creating object of Person class
Person1 = Person("John", 30)

# Accessing data using public method
Person1.display_info()  # Output: Name: John, Age: 30
```

```python
class BankAccount:

    # Constructor method → initializes the account balance
    def __init__(self, balance):
        self.__balance = balance  # Private attribute (data hiding)

    # Public method (Accessor / Getter)
    def get_balance(self):
        # Allows controlled access to private variable
        return self.__balance


# Creating object of BankAccount
account = BankAccount(1000)

# Accessing private data using public method
print(account.get_balance())  # Output: 1000
```

```python
class Person:

    # Constructor → initializes private attributes
    def __init__(self, name, age):
        self.__name = name  # Private attribute (Encapsulation)
        self.__age = age    # Private attribute (Encapsulation)

    # Public method to access private data
    def display_info(self):
        print(f"Name: {self.__name}, Age: {self.__age}")


# Create an instance of Person
person1 = Person("John", 30)
person1.display_info()  # Accessing private data via public method


# Child class Student inheriting from Person
class Student(Person):

    def __init__(self, name, age, student_id):
        # Calling parent class constructor using super()
        super().__init__(name, age)

        # student_id is not private here (normal attribute)
        self.student_id = student_id


# Create an instance of Student
student1 = Student("Alice", 20, "12345")  # Passing student_id

# Using inherited method from Person
student1.display_info()  # Displays name and age
```

```python
class BankAccount:

    # Constructor → initializes private balance
    def __init__(self, balance):
        self.__balance = balance  # Private attribute (data hiding)

    # Getter method → used to access private data
    def get_balance(self):
        return self.__balance

    # Method to deposit money
    def deposit(self, amount):
        # Check if amount is valid
        if amount > 0:
            self.__balance += amount  # Update private balance
            print("Amount deposited")
        else:
            print("Invalid deposit")

    # Method to withdraw money
    def withdraw(self, amount):
        # Check if sufficient balance is available
        if amount <= self.__balance:
            self.__balance -= amount  # Update private balance
            print("Withdrawal successful")
        else:
            print("Insufficient balance")


# Creating object of BankAccount
acc = BankAccount(8000)

# Depositing money
acc.deposit(900)

# Withdrawing money
acc.withdraw(600)

# Accessing final balance using getter method
print(acc.get_balance())  # Output: updated balance
```

```python
class Person:

    # Constructor → initializes private attribute
    def __init__(self, age):
        self.__age = age  # Private variable (data hiding)

    # Setter method → used to modify private data
    def set_age(self, age):
        # Validation check before updating
        if age > 0:
            self.__age = age  # Updating private variable
        else:
            print("Invalid age")  # Prevents wrong data

    # Getter method → used to access private data
    def get_age(self):
        return self.__age
```

```python
class Car:

    # Constructor → initializes private attribute (speed)
    def __init__(self, speed=0):
        self.__speed = speed  # Private variable (data hiding)

    # Method to increase speed
    def accelerate(self, amount):
        # Check if amount is valid
        if amount > 0:
            self.__speed += amount  # Increase speed safely

    # Method to decrease speed
    def brake(self, amount):
        # Ensure braking amount is valid and not more than current speed
        if 0 < amount <= self.__speed:
            self.__speed -= amount  # Decrease speed safely

    # Getter method → to access private speed
    def get_speed(self):
        return self.__speed


# Creating object of Car
car = Car()

# Increasing speed
car.accelerate(30)

# Accessing speed using getter method
print(car.get_speed())  # Output: 30
```
