# 🔹 Abstraction in Object-Oriented Programming (OOP)

Abstraction is a fundamental concept in OOP that focuses on **hiding the internal implementation details** and showing only the **essential features** of an object.

It helps us concentrate on **what an object does instead of how it does it**.

## 🔹 Key Points
- Hides complex logic from the user
- Improves code readability and maintainability
- Achieved using **abstract classes** and **abstract methods**
- In Python, abstraction is implemented using the `abc` module

## 🔹 Real-Life Example
A car 🚗:
- You can drive it using steering, brakes, and accelerator
- You don’t need to know how the engine works internally

👉 This is abstraction — using functionality without knowing internal details.

```python
from abc import ABC, abstractmethod  # Importing tools to create abstract classes

# Abstract Base Class (ABC)
class Animal(ABC):

    @abstractmethod  # This decorator makes the method abstract
    def make_sound(self):
        pass  # No implementation here → must be defined in child classes


# Dog class inherits from Animal
class Dog(Animal):

    def make_sound(self):
        # Providing implementation of abstract method
        print("Woof!")


# Cat class inherits from Animal
class Cat(Animal):

    def make_sound(self):
        # Providing implementation of abstract method
        print("Meow!")


# Creating object of Dog
dog = Dog()
dog.make_sound()  # Calls Dog's implementation

# Creating object of Cat
cat = Cat()
cat.make_sound()  # Calls Cat's implementation
```

```python
from abc import ABC, abstractmethod  # Importing Abstract Base Class tools

# Creating an abstract class
class Car(ABC):

    @abstractmethod  # Abstract method (must be implemented by child classes)
    def mileage(self):
        pass  # No implementation here

    # Concrete method (already implemented, can be used directly)
    def colour(self):
        print("White")  # Default color for all cars


# Child class Suzuki inheriting from Car
class Suzuki(Car):

    def mileage(self):
        # Providing implementation of abstract method
        print("The mileage is 25kmph")


# Child class Slavia inheriting from Car
class Slavia(Car):

    def mileage(self):
        # Providing implementation of abstract method
        print("The mileage is 30kmph")


# Creating object of Suzuki
suzuki = Suzuki()
suzuki.mileage()   # Calls Suzuki's mileage method
suzuki.colour()    # Calls common method from Car class

# Creating object of Slavia
slavia = Slavia()
slavia.mileage()   # Calls Slavia's mileage method
slavia.colour()    # Calls common method from Car class
```

```python
from abc import ABC, abstractmethod  # Importing Abstract Base Class tools

# Creating an abstract class
class Vehicle(ABC):

    @abstractmethod  # Abstract method (must be implemented by subclasses)
    def max_speed(self):
        pass  # No implementation here


# Child class Car inheriting from Vehicle
class Car(Vehicle):

    def max_speed(self):
        # Providing implementation of abstract method
        return 200  # Maximum speed of car


# Child class Bike inheriting from Vehicle
class Bike(Vehicle):

    def max_speed(self):
        # Providing implementation of abstract method
        return 100  # Maximum speed of bike


# Creating object of Car
car = Car()
print("The maximum speed of a car is: ", car.max_speed())
# Calls Car's implementation

# Creating object of Bike
bike = Bike()
print("The maximum speed of a bike is: ", bike.max_speed())
# Calls Bike's implementation
```

```python
from abc import ABC, abstractmethod  # Importing Abstract Base Class tools

# Creating an abstract class
class BankApp(ABC):

    # Concrete method (already implemented)
    def database(self):
        print('You are now Connected to the Database')

    @abstractmethod  # Abstract method (must be implemented in child class)
    def security(self):
        pass  # No implementation here


# Child class MobileApp inheriting from BankApp
class MobileApp(BankApp):

    def mobile_login(self):
        # Specific method for MobileApp
        print('Login into your Phone')

    def security(self):
        # Providing implementation of abstract method
        print('Mobile Security')


# Creating object of MobileApp
mob = MobileApp()

mob.database()       # Calls common method from BankApp
mob.mobile_login()   # Calls MobileApp specific method
mob.security()       # Calls implemented abstract method
```

```python
from abc import ABC, abstractmethod  # Importing Abstract Base Class tools

# Creating an abstract class
class Shape(ABC):

    @abstractmethod  # Abstract method (must be implemented by subclasses)
    def Area(self):
        pass  # No implementation here


# Child class Circle inheriting from Shape
class Circle(Shape):

    def __init__(self, radius):
        # Constructor method → used to initialize object attributes
        self.radius = radius  # Storing radius value

    def Area(self):
        # Providing implementation of abstract method
        return 3.14 * self.radius * self.radius  # Formula: πr²


# Creating object of Circle
circle = Circle(5)

# Calling Area method
print(circle.Area())  # Output: Area of circle
```
