### **Web Scraping**
It is the process of extracting data from websites using programs instead of manually copying information. In Python, it is commonly done using libraries like Requests and BeautifulSoup, which help download and parse webpage content. It is useful for collecting data such as text, tables, and statistics from the internet.

```python
#Using def, strip(), get_text(), text. Our objective is to create reusable scraping function and clean text.

import requests
from bs4 import BeautifulSoup

def scrape_quotes():
    url = "https://quotes.toscrape.com/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    quotes = soup.find_all('span', class_='text')
    authors = soup.find_all('small', class_='author')

    for quote, author in zip(quotes, authors):
        print(author.get_text(strip=True), "—",
              quote.get_text(strip=True))

scrape_quotes()
```

```python
#Web Scraping With Header. The purpose of using Header is because some websites block bots. Headers make requests look like a browser.

import requests
from bs4 import BeautifulSoup

url = "https://quotes.toscrape.com/"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

authors = soup.find_all('small', class_='author')

for a in authors[:5]:
    print(a.text)
```

```python
# Using enumerate() to add numbering while printing scraped data.
import requests
from bs4 import BeautifulSoup

url = "https://quotes.toscrape.com/"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

quotes = soup.find_all('span', class_='text')

for i, quote in enumerate(quotes[:5], 1):
    print(f"{i}. {quote.text}")
```

```python
# Scrape an HTML table and perform operations using Pandas DataFrame.

# Step 1 : Importing Libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Step 2 : Scrape the table
url = "https://www.worldometers.info/world-population/population-by-country/"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

table = soup.find("table")
rows = table.find_all("tr")
```

```python
# Step 3: Extract Data
data = []

for row in rows[1:]:   # skip header
    cols = row.find_all("td")

    country = cols[1].get_text(strip=True)
    population = cols[2].get_text(strip=True)

    data.append([country, population])
```

```python
# Step 4: Create DataFrame
df = pd.DataFrame(data, columns=["Country", "Population"])

print(df.head())
```

```python
# Step 5 : Table (DataFrame) Operations
# -> Check datatypes
print(df.dtypes)
# -> Convert population to numeric
df["Population"] = df["Population"].str.replace(",", "").astype(float)
# Select columns
df["Country"]
df[["Country", "Population"]]
```

```python
# Filter data
df[df["Country"] == "India"]
# Sort values
df.sort_values(by="Population", ascending=False)
```

```python
# Save file
df.to_csv("population.csv", index=False)
```

```python
# Step 6 : Basic analysis
print(df.describe())
```
