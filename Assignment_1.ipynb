```python
#Question - 1 : Create a Laptop class with a private attribute __price.
#Implement a method set_price(new_price) to update the price and get_price() to retrieve the price.

class Laptop:
    def __init__(self, price):
        self.__price = price

    def set_price(self, new_price):
        if new_price > 0:
            self.__price = new_price
            print("Price updated successfully.")
        else:
            print("Price must be greater than 0.")

    def get_price(self):
        return self.__price


# Creating object
laptop1 = Laptop(50000)
print("Original Price:", laptop1.get_price())
laptop1.set_price(55000)
print("Updated Price:", laptop1.get_price())
```

```python
#Question - 2: Create a BankAccount class with a private attribute __balance.
#Implement methods to deposit, withdraw, and check balance while ensuring the balance cannot be modified directly.

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
      if amount > 0:
        self.__balance += amount
        print("Amount deposited")
      else:
        print("Invalid deposit")

    def withdraw(self, amount):
      if amount <= self.__balance:
        self.__balance -= amount
        print("Withdrawal successful")
      else:
        print("Insufficient balance")

acc = BankAccount(8000)
acc.deposit(800)
acc.withdraw(300)
print(acc.get_balance())
```

```python
#Question - 3 : Create a Movie class with a private attribute __rating.
#Implement methods to set and get the rating, ensuring the rating is between 1 and 10.

class Movie:
    def __init__(self):
        self.__rating = 0

    def set_rating(self, rating):
        if 1 <= rating <= 10:
            self.__rating = rating
            print("Rating updated successfully.")
        else:
            print("Rating must be between 1 and 10.")

    def get_rating(self):
        return self.__rating


movie = Movie()
movie.set_rating(8.5)
print(f"The movie rating is: {movie.get_rating()}")
```

```python
#Question - 4 : Create a Book class with private attributes __title and __author.
#Implement a constructor to set these values and provide methods to retrieve them.

class Book:
    def __init__(self, title, author):
        self.__title = title
        self.__author = author

    def get_title(self):
        return self.__title

    def get_author(self):
        return self.__author


# Creating object
book1 = Book("Harry Potter", "J.K. Rowling")

print("Title:", book1.get_title())
print("Author:", book1.get_author())
```

```python
#Question - 5 : Create a Car class with a private attribute __speed.
#Implement methods to increase speed, decrease speed, and get the current speed, ensuring speed does not go below 0.

class Car:
    def __init__(self):
        self.__speed = 0

    def increase_speed(self, value):
        if value > 0:
            self.__speed += value
            print("Speed increased to:", self.__speed)
        else:
            print("Increase value must be positive.")

    def decrease_speed(self, value):
        if value > 0:
            self.__speed -= value
            if self.__speed < 0:
                self.__speed = 0
            print("Speed decreased to:", self.__speed)
        else:
            print("Decrease value must be positive.")

    def get_speed(self):
        return self.__speed


# Creating object
car = Car()

car.increase_speed(100)
car.decrease_speed(30)
print("Current Speed:", car.get_speed())
```
